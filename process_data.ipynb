```python
from PIL import Image
import os

#define the target size for images
target_size = (224, 224)

#fnction to process images
def process_images(input_path, output_path, target_size):
    for root, dirs, files in os.walk(input_path):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            try:
                #open and process the image
                img = Image.open(file_path).convert('RGB')
                img_resized = img.resize(target_size, Image.Resampling.LANCZOS)

                #save the resized image
                save_dir = os.path.join(output_path, os.path.relpath(root, input_path))
                os.makedirs(save_dir, exist_ok=True)

                save_path = os.path.join(save_dir, os.path.splitext(file_name)[0] + ".jpg")
                img_resized.save(save_path, "JPEG")
                print(f"Processed and saved: {save_path}")
            except Exception as e:
                print(f"Error processing file {file_path}: {e}")

input_path = "datasets/Danger Of Extinction"  #path where the dataset is stored
output_path = "processed_images"  #path to save the processed images

process_images(input_path, output_path, target_size)
```
